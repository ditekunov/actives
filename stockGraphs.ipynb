```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
def zerolistmaker(n):
    listofzeros = [0] * n
    return listofzeros

def targetMaker(n):
    listofTarget = [5000] * n
    return listofTarget

totalList = [690, 1143, 1209, 1214, 1222, 1228, 1225, 1636, 1640, 1649, 1653, 1643, 1642, 1637]

incomesList = [-1, 4, 24, 30, 32, 39, 38, 39, 45, 54, 58, 47, 49, 46]

days = list(range(0, len(totalList)))

zeroes = zerolistmaker(len(totalList))

target = targetMaker(len(totalList))

obsoleteValue = list()

bankBalance = 500

for i in totalList:
    obsoleteValue.append(i + bankBalance)
```

```python
plt.title('Total sum')
sns.pointplot(days, totalList , color="blue")
sns.pointplot(days, target, markers="")
```

```python
plt.title('Income')
sns.pointplot(days, incomesList , color="blue")
sns.pointplot(days, zeroes, markers="")
```

```python
trueIncomesList = list()

for i in range(0, len(totalList)):
    trueIncomesList.append(incomesList[i] / totalList[i])
    
print(trueIncomesList)
```

```python
plt.title('True income')
sns.pointplot(days, trueIncomesList , color="blue")
sns.pointplot(days, zeroes, markers="")
```

```python
plt.title('Obsolete sum')
sns.pointplot(days, obsoleteValue, color="blue")
sns.pointplot(days, target, markers="")
```
